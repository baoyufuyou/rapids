```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
resultFilename = 'local_cluster_experiment_results.csv'
resultsDF = pd.read_csv( resultFilename )
```

```python
resultsDF.head()
```

```python
resultsDF.columns
```

```python
plt.figure( figsize = (13,5))
tickLabels = list( map( str, resultsDF['nSamples'] ) )
#tickLabels = list( resultsDF['nSamples'] )
xTicks = range( len( resultsDF['elapsedTime'] ) )
plt.bar( x = xTicks,
        height = resultsDF['elapsedTime'],
        tick_label = tickLabels )
plt.xticks(rotation=70)
plt.show()
```

```python
groupedDF = resultsDF.groupby(['nSamples', 'nGPUs'])
```

```python
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'xx-large',
         'axes.titlesize':'xx-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
```

```python
fig = plt.figure( figsize = (25,10));
ax = plt.subplot(111)
ax = groupedDF['elapsedTime'].mean().plot.bar( zorder=3, 
                                               color=[118/255., 185/255., 0.],
                                              linewidth=1.5, edgecolor='darkgreen'
                                             );
ax.grid(axis='y', which='both', zorder=0)
plt.ylabel('Elapsed Time',labelpad=20)
plt.xlabel('( nSamples, nGPUs )', labelpad=20)
plt.title('Scale Up Performance on Synthetic Dataset', pad=20);
```
